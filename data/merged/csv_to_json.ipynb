```python
import os
import pandas as pd
import json
import numpy as np
cwd = os.getcwd()
parentwd = os.path.dirname(cwd)
```

```python
city = pd.read_csv(os.path.join(cwd,"city_level.csv"))
uni = pd.read_csv(os.path.join(cwd,"university_level.csv"))
```

```python

# Creating university database
uni_db = {}

for _, row in uni.iterrows():
    university_id = row["university_id"]
    country_code = row["country_code"]
    
    # University data structure
    uni_db[university_id] = {
        "name": row["university"],
        "website": "",  # Placeholder, as no website avaiable for now
        "location": {
            "city": row["city"],
            "countryCode": country_code,
        },
        "qsRankingInfo": {
            "rank": row["2025_rank"],
            "academic_reputation": row["academic_reputation"],
            "international_students": row["international_students"],
            "employment_outcomes": row["employment_outcomes"],
            "sustainability": row["sustainability"],
            "qs_overall_score": row["qs_overall_score"]
        },
        "tuitionFee": {
            "amount": row["tuition_fee_yearly_usd"],
            "provinance": row["method"], 
        },
    }
```

```python

# Creating city database
city_db = {}
country_db = {}
for _, row in city.iterrows():
    city_name = row["city"]
    country_code = row["country_code"]

    city_db[city_name] = {
            "name": row["city"],  
            "country": row["country"],
            "country_code": country_code,
            "temperature": row["temp_year_avg"],
            "universities": [],
            "costOfLiving": {
                "cost_of_living_index": row["cost_of_living_index"],
                "rent_index": row["rent_index"],
                "cost_of_living_plus_rent_index": row["cost_of_living_plus_rent_index"],
                "groceries_index": row["groceries_index"],
                "restaurant_price_index": row["restaurant_price_index"],
                "local_purchasing_power_index": row["local_purchasing_power_index"],
            },
            "efScore": {
                "ef_score": row["ef_score"],
                "ef_level": row["ef_level"]
            },  
        }

    # Country-level data structure
    if country_code not in country_db:
        country_db[country_code] = {
            "name": row["country"], 
            "temperature": row["temp_year_avg"],
            "universities": [],
            "costOfLiving": {
                "cost_of_living_index": row["cost_of_living_index"],
                "rent_index": row["rent_index"],
                "cost_of_living_plus_rent_index": row["cost_of_living_plus_rent_index"],
                "groceries_index": row["groceries_index"],
                "restaurant_price_index": row["restaurant_price_index"],
                "local_purchasing_power_index": row["local_purchasing_power_index"],
            },
            "efScore": {
                "ef_score": row["ef_score"],
                "ef_level": row["ef_level"]
            },  
        }

```

```python
city_db
```

```python
for _, row in uni.iterrows():
    university_id = row["university_id"]
    country_code = row["country_code"]
    city_name = row["city"]

    if country_code == "UK":
        country_code = "GB"
    if country_code not in country_db:
        continue
    country_db[country_code]["universities"].append(university_id)
    if city_name in city_db:
        city_db[city_name]["universities"].append(university_id)
    
    
# Combine both databases again
final_data = {"uni_db": uni_db, "country_db": country_db,"city_db": city_db}

```

```python
# Save the updated JSON
import simplejson
json_file_path = os.path.join(cwd,"json_db.json")
with open(json_file_path, "w") as json_file:
    simplejson.dump(final_data, json_file, ignore_nan=True, sort_keys=True, indent=4 * ' ')

# Return the new JSON file path
json_file_path
```
